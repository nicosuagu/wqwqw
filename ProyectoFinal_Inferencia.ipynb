**A. Distribución de un estimador máximo verosímil cuando no hay
condiciones de regularidad.**


Considere una muestra aleatoria $X_1,X_2,...,X_n$ de una distribución $U [0, \theta=12 ] $
, es
decir, donde el verdadero valor del parámetro es igual al número del grupo. Se
sabe que
$\hat{θ}_{MLE}=X^{(n)}$
, sin embargo, como no se tienen condiciones de regularidad,
este estimador no tiene una distribución asintótica normal (cuando es debidamente
normalizado). Vamos a estudiar también la convergencia en distribución de la
variable aleatoria $R_n=n(θ-X^{(n)})$

a. Genere m =1000 simulaciones de una muestra aleatoria de tamaño n =10
de la distribución $U [0, \theta=12 ] $, usando la función runif de R o su equivalente en otro lenguaje de programación. Para cada una de las
m muestras, calcule y
almacene la estimación máximo-verosímil del parámetro. Haga un histograma
de las m=1000 estimaciones y superponga sobre este histograma la función de
densidad teórica de X^{(n)}.

```r
%%R

# Parámetros
m <- 1000   # Número de simulaciones
n <- 10     # Tamaño de la muestra
theta <- 12 # Parámetro de la distribución uniforme

# Generar las m muestras y estimar el MLE
estimaciones <- rep(0, m) # Vector para almacenar las estimaciones

for (i in 1:m) {
  muestra <- runif(n, 0, theta)  # Generar una muestra aleatoria
  estimaciones[i] <- max(muestra)  # Estimar el MLE
}

# Crear el histograma de las estimaciones
hist(estimaciones, freq = FALSE, breaks = "FD", main = "Histograma del MLE", xlab = "Estimación Máximo Verosímil")
hx<-density(estimaciones)
lines(hx, col="red") #Agregamos la linea de densidad del MLE 
```

Este fenómeno se explica por el teorema de la convergencia en probabilidad, que establece que cuando el tamaño de la muestra tiende a infinito, el estimador máximo verosímil converge en probabilidad al verdadero valor del parámetro. En otras palabras, a medida que la muestra se hace más grande, es más probable que el estimador MLE esté cerca del verdadero valor del parámetro.

En el caso de la distribución uniforme en el intervalo [0, 12], cuando el tamaño de la muestra aumenta, el estimador MLE (que es el máximo de la muestra) tiende a acercarse al valor $\theta=12$. Esto se debe a que, a medida que se tienen más observaciones, es más probable que la muestra contenga valores cercanos al extremo superior del intervalo, lo que hace que el máximo de la muestra se acerque a ese valor.

Por lo tanto, a medida que el tamaño de la muestra aumenta, las realizaciones del estimador máximo verosímil tienden a concentrarse alrededor del valor verdadero del parámetro, en este caso, $\theta=12$.

```r
#@title
%%R
m <- 1000  # Número de muestras
n_values <- c(10, 50, 100, 200, 500, 1000)  # Valores de n

theta <- 12  # Valor verdadero del parámetro

set.seed(123)  # Semilla para reproducibilidad

# Función para calcular R_n
calculate_Rn <- function(sample, theta) {
  n <- length(sample)
  Rn <- n * (theta - max(sample))
  return(Rn)
}

# Almacenar los resultados
results <- matrix(NA, nrow = m, ncol = length(n_values))

# Generar las muestras y calcular R_n
for (i in 1:length(n_values)) {
  n <- n_values[i]
  for (j in 1:m) {
    sample <- runif(n, min = 0, max = theta)
    results[j, i] <- calculate_Rn(sample, theta)
  }
}

# Histogramas de los valores de R_n para cada n
par(mfrow = c(3, 2))  # Dividir la ventana gráfica en una cuadrícula de 3x2

for (i in 1:length(n_values)) {
  n <- n_values[i]
  hist(results[, i], main = paste("Histograma de R_n (n =", n, ")"), xlab = "R_n", prob = TRUE)
  curve(dnorm(x, mean = 0, sd = sqrt(n/12)), add = TRUE, col = "red")  # Curva de distribución normal
}

```

e. Pruebe formalmente su conjetura del apartado anterior, calculando el límite en
distribución cuando $n$ tiende a infinito:
$R_n=n(θ-\hat{θ}_{(MLE)})$ a que tiende en distribución?

Nota: El estimador ML de este modelo es súper consistente y por eso su factor
de escala no es $raiz(n)$ sino n

La distribución límite de la variable aleatoria $R_n = n(\theta - \hat{\theta}_{MLE})$ cuando $n$ tiende a infinito, podemos utilizar la función de distribución acumulativa (CDF) y observar hacia qué distribución converge.

En este caso, el estimador de máxima verosimilitud $\hat{\theta}_{MLE}$ es igual al máximo de la muestra, $X^{(n)}$. Dado que tenemos una distribución uniforme $U[0, \theta]$, el valor máximo de la muestra $X^{(n)}$ se acerca al verdadero valor del parámetro $\theta$ a medida que $n$ aumenta.

Entonces, podemos reescribir la variable aleatoria $R_n$ como $R_n = n(\theta - X^{(n)})$. La función de distribución acumulativa de $R_n$ se define como:

$$F_{R_n}(x) = P(R_n \leq x) = P(n(\theta - X^{(n)}) \leq x)$$

Dividimos por $n$ y multiplicamos por $-1$ para cambiar la dirección de la desigualdad:

$$F_{R_n}(x) = P(X^{(n)} \geq \theta - \frac{x}{n})$$

La probabilidad de que el máximo de la muestra $X^{(n)}$ sea mayor o igual a $\theta - \frac{x}{n}$ es igual a la probabilidad de que todos los elementos de la muestra sean mayores o iguales a $\theta - \frac{x}{n}$:

$$F_{R_n}(x) = [P(X_1 \geq \theta - \frac{x}{n})]^n = \left(\frac{\theta - \frac{x}{n}}{\theta}\right)^n = \left(1 - \frac{x}{n\theta}\right)^n$$

Cuando $n$ tiende a infinito, podemos tomar el límite de esta función de distribución acumulativa:

$$\lim_{n\to\infty} F_{R_n}(x) = \lim_{n\to\infty} \left(1 - \frac{x}{n\theta}\right)^n$$

Este límite converge hacia la función de distribución acumulativa de una distribución exponencial con parámetro $\frac{1}{\theta}$:

$$\lim_{n\to\infty} F_{R_n}(x) = e^{-\frac{x}{\theta}}$$

Por lo tanto, la distribución límite de la variable aleatoria $R_n$ cuando $n$ tiende a infinito es una distribución exponencial con parámetro $\frac{1}{\theta}$.


## **B. Cálculo de una estimación máximo-verosímil cuando no hay una solución analítica**

a. Justifique, de manera analítica, que los siguientes resultados son válidos para
una variable aleatoria, X, con distribución




\begin{equation*}
X \sim \text{Cauchy}(\theta, 1)
\end{equation*}


$\textbf{I. Valor esperado:}$

El valor esperado (o media) de una variable aleatoria se calcula como la integral de $x$ multiplicada por la función de densidad de probabilidad. En el caso de la distribución de Cauchy, la función de densidad de probabilidad es:

$[
f(x; \theta) = \frac{1}{\pi} \cdot \frac{1}{1 + (x - \theta)^2}
]$

La integral de $x$ multiplicada por $f(x)$ en el rango de $-\infty$ a $\infty$ puede ser dividida en dos integrales:

$[
E[X] = \int_{-\infty}^{ \theta} x \cdot f(x; \theta) \, dx + \int_{ \theta}^{\infty} x \cdot f(x; \theta) \, dx
]$

Ambas integrales no convergen a un valor finito. Esto se debe a las colas largas y pesadas de la distribución de Cauchy, que provocan que los valores extremos tengan una influencia significativa en el cálculo del valor esperado. Por lo tanto, en la distribución de Cauchy, el valor esperado no existe.

$\textbf{II. Mediana poblacional:}$

La mediana poblacional, también conocida como percentil 0.5, se puede obtener al encontrar el valor de $x$ para el cual la función de distribución acumulativa (CDF) alcanza $0.5$. La función de distribución acumulativa de la distribución de Cauchy es:

$[
F(x; \theta) = \frac{1}{2} + \frac{1}{\pi} \cdot \text{atan}(x - \theta)
]$

Para encontrar la mediana, igualamos $F(x; \theta)$ a $0.5$ y resolvemos para $x$:

$[
\frac{1}{2} + \frac{1}{\pi} \cdot \text{atan}(x - \theta) = 0.5
]$

$[
\frac{1}{\pi} \cdot \text{atan}(x - \theta) = 0
]$

$[
\text{atan}(x - \theta) = 0
]$

$[
x - \theta = 0
]$

$[
x = \theta
]$

Por lo tanto, la mediana poblacional o percentil 0.5 de la distribución de Cauchy es igual a $\theta$.

$\textbf{III. Moda poblacional:}$

La moda poblacional se puede determinar encontrando el valor de $x$ para el cual la función de densidad de probabilidad alcanza su máximo. Para la distribución de Cauchy, la función de densidad de probabilidad es:

$[
f(x; \theta) = \frac{1}{\pi} \cdot \frac{1}{1 + (x - \theta)^2}
]$

Derivando esta función con respecto a $x$ e igualando a cero, podemos encontrar la moda:

$[
\frac{d}{dx} [f(x; \theta)] = \frac{d}{dx} \left[\frac{1}{\pi} \cdot \frac{1}{1 + (x - \theta)^2}\right] = 0
]$

Resolviendo esta ecuación, encontramos:

$[
  \frac{2(x - \theta)}{\pi((x - \theta)^2+1)^2} = 0
]$
$[
  2(x - \theta) = 0
]$
$[
    \theta = x
]$



En resumen, para la distribución de Cauchy:

$\textbf{I.} $El valor esperado no existe.

$\textbf{II.}$ La mediana poblacional o percentil 0.5 es igual a $\theta$.

$\textbf{III.}$ La moda poblacional es igual a $\theta$.

```r
%%R
m <- 1000  # Número de simulaciones
n <- 10  # Tamaño de la muestra


# Definir la función de log-verosimilitud de la distribución de Cauchy
log_likelihood_cauchy <- function(data, location) {
  n <- length(data)
  log_likelihood <- -n * log(pi) - sum(log(1 + ((data - location) )^2))
  return(log_likelihood)
}


set.seed(123)  # Establecer semilla para reproducibilidad

estimaciones_mle <- vector("numeric", m)  # Vector para almacenar las estimaciones MLE
estimaciones_analogia <- vector("numeric", m)  # Vector para almacenar las estimaciones por analogía

for (i in 1:m) {
  # Generar una muestra de datos de ejemplo de una distribución de Cauchy
 
  muestra <- rcauchy(n, location = 12, scale = 1)
 
  # Función objetivo para maximizar la log-verosimilitud
  objetivo <- function(params) {
    location <- params[1]
    return(-log_likelihood_cauchy(muestra, location))
  }
 
  # Estimación de MLE utilizando optimización numérica
  resultado <- optim(c(0, 1), objetivo)  # Valores iniciales para la optimización
 
 
 
  estimaciones_mle[i] <- resultado$par[1]  # Estimación máximo-verosímil del parámetro
  estimaciones_analogia[i] <- median(muestra)  # Estimación por analogía (mediana)
}

# Histograma de las estimaciones MLE
hist(estimaciones_mle, main = "Estimaciones MLE", xlab = "Estimación", col = "lightblue", breaks = 40,xlim = c(0, 20))
abline(v = 12, lwd = 2, col = "red")  # Línea vertical para el verdadero valor del parámetro
abline(v = mean(estimaciones_mle), lwd = 2, col = "green")  # Línea vertical para el promedio de las estimaciones
# Personalizar los marcadores y etiquetas del eje x
axis(1, at = seq(0, 20, by = 3), labels = seq(0, 20, by = 3))

# Histograma de las estimaciones por analogía
hist(estimaciones_analogia, main = "Estimaciones por Analogía", xlab = "Estimación", col = "lightblue")
abline(v = 12, lwd = 2, col = "red")  # Línea vertical para el verdadero valor del parámetro
abline(v = mean(estimaciones_analogia), lwd = 2, col = "green")  # Línea vertical para el promedio de las estimaciones

# Tabla de promedio y varianza de las estimaciones
tabla <- data.frame(
  Estimador = c("MLE", "Analogía"),
  Promedio = c(mean(estimaciones_mle), mean(estimaciones_analogia)),
  Varianza = c(var(estimaciones_mle), var(estimaciones_analogia))
)
print(tabla)
```

```r
%%R
n<- 1000
m <- 1000  # Número de simulaciones
# Definir la función de log-verosimilitud de la distribución de Cauchy
log_likelihood_cauchy <- function(data, location) {
  n <- length(data)
  log_likelihood <- -n * log(pi) - sum(log(1 + ((data - location) )^2))
  return(log_likelihood)
}


set.seed(123)  # Establecer semilla para reproducibilidad

estimaciones_V <- vector("numeric", m)  # Vector para almacenar las estimaciones V
estimaciones_W <- vector("numeric", m)  # Vector para almacenar las estimaciones por W

for (i in 1:m) {
  # Generar una muestra de datos de ejemplo de una distribución de Cauchy
  
  muestra <- rcauchy(1000, location = 12, scale = 1)
  
  # Función objetivo para maximizar la log-verosimilitud
  objetivo <- function(params) {
    location <- params[1]
    return(-log_likelihood_cauchy(muestra, location))
  }
  
  # Estimación de MLE utilizando optimización numérica
  resultado <- optim(c(0, 1), objetivo)  # Valores iniciales para la optimización
  
  
  
  estimaciones_V[i] <- 100*(resultado$par[1]-12)  # Estimación máximo-verosímil del parámetro
  estimaciones_W[i] <- 100*(median(muestra)-12)  # Estimación por analogía (mediana)
}

media <- 0
desviacion <- 2.467


par(mfrow = c(1, 2))
# Histograma de las estimaciones V
y <-hist(estimaciones_V, main = "Estimaciones V", xlab = "Estimación", col = "lightblue", breaks = 40,xlim = c(0, 20))
abline(v = mean(estimaciones_V), lwd = 2, col = "green")  # Línea vertical para el promedio de las estimaciones
# Personalizar los marcadores y etiquetas del eje x
axis(1, at = seq(0, 20, by = 3), labels = seq(0, 20, by = 3))

# Histograma de las estimaciones por W
x <-hist(estimaciones_W, main = "Estimaciones por W", xlab = "Estimación", col = "lightblue")
curve(dnorm(x, mean = media, sd = desviacion), from = -15, to = 15, add = TRUE, col = "red", lwd = 2)

# Línea vertical para el promedio de las estimaciones
abline(v = mean(estimaciones_W), lwd = 2, col = "green")

```

```r
%%R
tabla <- data.frame(
  Estimador = c("V", "W"),
  Promedio = c(mean(estimaciones_V), mean(estimaciones_W)),
  Varianza = c(var(estimaciones_V), var(estimaciones_W))
)
print(tabla)

```

parece ser que V tiende a tener menor varianza y ser menos segado que W luego resulta ser mas efectivo en error cuadratico medio

## **C. Comparación de varios estimadores por el método de los momentos.**


Sea una muestra aleatoria $X_1$, $X_2$,...,$X_n$ de una distribución *Pois*(𝛳=12).

Vamos a construir el primer $\hat{θ}_{MOM,1}$ , segundo $\hat{θ}_{MOM,2}$ y tercer $\hat{θ}_{MOM,3}$ estimador por el método de los momentos. 

Entonces, sabemos que para una distribución *Poisson* se tiene que E(X) = θ, y que Var(X)=θ. 

El primer momento muestral se define como:

$\bar{X}=\frac{1}{n}
\sum_{i=1}^{n}x_{i} $

Por lo tanto, por el método de los momentos se tiene que $\hat{θ}=\bar{X}=\hat{θ}_{MOM,1}$.

Ahora, para una distrubición *Poisson*, la varianza es el segundo momento informativo. Por lo tanto, $\hat{θ}_{MOM,2}=Var(x)$ 

y por último, para el tercer momento tenemos que:

$\hat{θ}_{MOM,3}=\frac{1}{n}
\sum_{i=1}^{n}(x_{i}-\bar{x})^3$ 

```r
#@title
%%R
m <- 1000 
n <- 10 
lambda <- 12 
simulaciones <- matrix(rpois(m * n, lambda), ncol = n)

estimador_1 <- numeric(m)
estimador_2 <- numeric(m)
estimador_3 <- numeric(m)
for (i in 1:m) {
muestra <- simulaciones[i,]


estimador_1[i] <- mean(muestra)
 

estimador_2[i] <- var(muestra)
 

estimador_3[i] <- mean((muestra - mean(muestra))^3)
}

#Histogramas

# Estimador #1
hist(estimador_1, main = "Histograma estimador 1 con n=10", xlab = "Estimador 1")
abline(v = lambda, col = "black", lwd = 2)  
abline(v = mean(estimador_1), col = "Deeppink", lwd = 2)  

# Estimador #2
hist(estimador_2, main = "Histograma estimador 2 con n=10", xlab = "Estimador 2")
abline(v = lambda, col = "coral1", lwd = 2)
abline(v = mean(estimador_2), col = "blue", lwd = 2)

# Estimador #3
hist(estimador_3, main = "Histograma estimador 3 con n=10", xlab = "Estimador 3")
abline(v = lambda, col = "darkorchid1", lwd = 2)
abline(v = mean(estimador_3), col = "orange", lwd = 2)
```

Teniendo en cuenta que si el sesgo de un estimador $\hat{θ}(X)$ para un parámetro θ es igual a cero, es decir que si el promedio de sus estimaciones es igual a el valor del parámetro θ, decimos que el estimador es insesgado. 

En nuestro caso en especifico, aunque ninguno de los tres estimadores es exactamente igual a el valor del parámetro, si se puede ver que estan muy cerca, especialmente mirando el estimador 1 y 2, lo que nos podria dar indicios de que los tres pueden ser buenos estimadores, pero que podemos revisar en términos de MSE. 

```r
#@title
%%R
# Promedio y varianza de las estimaciones
promedio_estimador_1 <- mean(estimador_1)
varianza_estimador_1 <- var(estimador_1)

promedio_estimador_2 <- mean(estimador_2)
varianza_estimador_2 <- var(estimador_2)

promedio_estimador_3 <- mean(estimador_3)
varianza_estimador_3 <- var(estimador_3)

# Tabla de resultados
resultados <- data.frame(
  Estimador = c("Estimador 1", "Estimador 2", "Estimador 3"),
  Promedio = c(promedio_estimador_1, promedio_estimador_2, promedio_estimador_3),
  Varianza = c(varianza_estimador_1, varianza_estimador_2, varianza_estimador_3)
)

print(resultados)
```

Aqui podemos ver que para el estimador 1 y el estimador 2, el promedio es muy cercano a el valor real del parámetro. Recordemos que un buen estimador debe ser preciso (poca varianza) y exacto (poco o sesgo nulo). 

Para confirmar que en efecto, el mejor estimador sería el estimador 1, es decir el promedio muestral, podemos analizarlos en términos de los valores de error cuadrático medio.

```r
#@title
%%R
# MSE
mse_estimador_1 <- mean(((estimador_1)-lambda)^2)

mse_estimador_2 <- mean(((estimador_2)-lambda)^2)

mse_estimador_3 <- mean(((estimador_3)-lambda)^2)

# Tabla de resultados con MSE
resultados1 <- data.frame(
  Estimador = c("Estimador 1", "Estimador 2", "Estimador 3"),
  Promedio = c(promedio_estimador_1, promedio_estimador_2, promedio_estimador_3),
  Varianza = c(varianza_estimador_1, varianza_estimador_2, varianza_estimador_3),
  MSE = c(mse_estimador_1, mse_estimador_2, mse_estimador_3)
)

print(resultados1)
```

Por lo tanto, podemos ver que de "mejor" a "peor" estimador, para este caso, tenemos el siguiente orden:
1. $\hat{θ}_{MOM,1}$ 
2. $\hat{θ}_{MOM,2}$ 
3. $\hat{θ}_{MOM,3}$

## *D. Técnica de remuestreo (Bootstrap)*

En ejercicios de simulación, cuando es necesario calcular algunas propiedades de una
variable aleatoria, poder generar muestras de la verdadera distribución de los datos es
de gran utilidad para verificar que los resultados teóricos se tienen. Sin embargo, en la
vida real, rara vez se conoce la “verdadera” distribución de los datos y solo se cuenta
con un conjunto de datos $x_1,x_2,...;x_n$ recogidos.

Ahora bien, uno podría hacer un histograma de los datos, suponer un modelo para
ellos a partir de la evidencia del histograma y luego, estimar los correspondientes
parámetros. Sin embargo, es posible que la elección de ese modelo “adecuado” no
sea tan sencilla y quisiéramos poder contar con un procedimiento que nos permitiera
estudiar a una determinada estadística SOLO con la información que tienen los datos,
sin necesidad de recurrir a hacer suposiciones sobre su distribución.
Considere entonces una muestra aleatoria
$x_1,x_2,...;x_n$ de una distribución $N(0,σ^2)$
. Supongamos que el interés que se tiene es estimar a $σ$. Se proponen,

entonces, los siguientes estimadores:



a. ¿Qué propiedades puede resaltar de $S_n$ para estimar a σ? Justifique.

b. Determine el valor de la constante $c$ en $D_n$ para que este estimador sea insesgado. ¿Qué otras propiedades deseables son fácilmente verificables para
$D_n$?


$$D_n=\frac{c}{n} \sum_{i=1}^{n}|X_i| $$
Sesgo
$$ E[D_n] -\sigma$$
$$ E[\frac{c}{n} \sum_{i=1}^{n}|X_i|]- \sigma$$
$$ \frac{c}{n} \sum_{i=1}^{n}E[|X_i|]- \sigma$$
Se calcula el $ E[|X_i|]$

$$ E[|X_i|]=\int_{-\infty}^{\infty}|X|\frac{1}{\sqrt{2\pi}\sigma} e^{-x^2/2\sigma^2} dx $$

Se puede ver que la función de la integral es par ya que $f(x)=f(-x)$

$$ |X|\frac{1}{\sqrt{2\pi}\sigma} e^{-x^2/2\sigma^2} =|-X|\frac{1}{\sqrt{2\pi}\sigma} e^{-x^2/2\sigma^2}  $$

Luego, por propiedades de las integrales se puede reescribir:

$$\int_{-\infty}^{\infty}f(x)=2\int_{0}^{\infty}f(x)$$
$$ E[|X_i|]=2\int_{0}^{\infty}|X|\frac{1}{\sqrt{2\pi}\sigma} e^{-x^2/2\sigma^2} dx $$

Como en el intervalo solo hay x mayor o igual a $0$, se puede quitar el valor absoluto, quedando:

$$ E[|X_i|]=2\int_{0}^{\infty} \frac{x}{\sqrt{2\pi}\sigma} e^{-x^2/2\sigma^2} dx $$
Se saca la constante 
$$ E[|X_i|]=\frac{2}{\sqrt{2\pi}\sigma} \int_{0}^{\infty} x e^{-x^2/2\sigma^2} dx $$
Hacemos una sustitución
$$v=-\frac{u}{\sigma} $$
$$dv=-\frac{1}{\sigma} du$$
$$ du=-\sigma dv$$

$$ E[|X_i|]=\frac{2}{\sqrt{2\pi}\sigma} \int_{0}^{\infty} -\sigma e^{v} dv $$
$$= \frac{-2\sigma}{\sqrt{2\pi}\sigma} \int_{0}^{\infty}  e^{v} dv $$
$$= \frac{-2\sigma}{\sqrt{2\pi}\sigma}  e^{v} $$

Reescribimos 

$$v=\frac{-u}{\sigma} $$
$$u=\frac{x^2}{2\sigma} $$

$$= \frac{-2\sigma}{\sqrt{2\pi}\sigma}  e^{\frac{-u}{\sigma}} $$
$$= \frac{-2\sigma}{\sqrt{2\pi}\sigma}  e^{\frac{-x^2}{2\sigma^2}}|_0^\infty $$

$$= \frac{-2\sigma}{\sqrt{2\pi}\sigma} (\lim_{t \infty} e^{\frac{-t}{2\sigma^2}}- e^0) $$
$$= \frac{-2\sigma}{\sqrt{2\pi}\sigma} (0-1) $$
$$= \frac{2\sigma}{\sqrt{2\pi}\sigma}  $$
$$=\sqrt{\frac{2}{\pi}} \sigma$$

Entonces
$$E[|X_i|]=\sqrt{\frac{2}{\pi}} \sigma$$
Calculando nuevamente el sesgo 
$$ \frac{c}{n} \sum_{i=1}^{n}E[|X_i|]- \sigma$$
$$ \frac{c}{n} n(\sqrt{\frac{2}{\pi}} \sigma)- \sigma$$
$$ c(\sqrt{\frac{2}{\pi}} \sigma)- \sigma$$

Para que sea insesgado, esta operación debe ser igual a 0, luego
$$ c(\sqrt{\frac{2}{\pi}} \sigma)- \sigma=0$$
$$ \sigma(c(\sqrt{\frac{2}{\pi}} )- 1)=0$$
$$ c(\sqrt{\frac{2}{\pi}} )- 1=0$$
$$ c(\sqrt{\frac{2}{\pi}} )= 1$$
$$ c=\frac{1}{\sqrt{\frac{2}{\pi}}}=\sqrt{\frac{\pi}{2}}$$

b. ¿Qué sucede con la variabilidad de cada estimador a medida que
aumenta el tamaño de muestra? La variabilidad viene representada por
la diferencia entre el percentil 95% y el percentil 5%.

```r

#generamos un vector con los percentiles que guardamos
#tamaño de los datos
n <- 1000 # Número de simulaciones
m_values <- c(50, 100, 200, 500, 1000) # Valores de m a probar

# Crear vectores para almacenar los resultados
promedioSN <- numeric(length(m_values))
percentil5_SN <- numeric(length(m_values))
percentil95_SN <- numeric(length(m_values))
promedioDN <- numeric(length(m_values))
percentil5_DN <- numeric(length(m_values))
percentil95_DN <- numeric(length(m_values))

for (i in 1:length(m_values)) {
  m <- m_values[i] # Tamaño de la muestra
  
  estimacionSN <- c(1:n) # Vector para guardar las estimaciones de Sn
  estimacionDN <- c(1:n) # Vector para guardar las estimaciones de Dn
  for (j in 1:n) {
    muestra <- rnorm(m, 0, sqrt(12))
    estimacionSN[j] <- sd(muestra)
    estimacionDN[j] <- (sqrt(pi/2)/m) * sum(abs(muestra))
  }
  
  # Calcular promedio y percentiles para cada estimador
  promedioSN[i] <- mean(estimacionSN)
  percentil5_SN[i] <- quantile(estimacionSN, 0.05)
  percentil95_SN[i] <- quantile(estimacionSN, 0.95)
  promedioDN[i] <- mean(estimacionDN)
  percentil5_DN[i] <- quantile(estimacionDN, 0.05)
  percentil95_DN[i] <- quantile(estimacionDN, 0.95)
}

# Gráfico de evolución del promedio, el percentil 5% y el percentil 95%
par(mfrow = c(1, 1))  
plot(m_values, promedioSN, type = "b", col = "red", xlab = "Tamaño de muestra (m)", ylab = "Valor", 
     main = "Evolución de Promedio, Percentil 5% y Percentil 95%", ylim = range(c(promedioSN, percentil5_SN, percentil95_SN, promedioDN, percentil5_DN, percentil95_DN)))
lines(m_values, percentil5_SN, type = "b", col = "blue")
lines(m_values, percentil95_SN, type = "b", col = "green")
lines(m_values, promedioDN, type = "b", col = "purple")
lines(m_values, percentil5_DN, type = "b", col = "orange")
lines(m_values, percentil95_DN, type = "b", col = "brown")
plot.new()
legend("top", legend = c("Promedio SN", "Percentil 5% SN", "Percentil 95% SN", "Promedio DN", "Percentil 5% DN", "Percentil 95% DN"), 
       col = c("red", "blue", "green", "purple", "orange", "brown"), lty = 1, cex = 1)
```
